```python
# 1. Import the numpy and pandas libraries
import pandas as pd
import numpy as np
```

```python
# 2. Load the iris dataset
df = pd.read_csv("iris.csv")
df.head()
```

```python
# 3. Seprate the dataset into input and output

input_features = df.iloc[:,:-1]
output_feature = df.iloc[:,-1]
```

```python
print(input_features)
print(output_feature)
```

```python
# 4. Devide the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_features, output_feature, test_size=.20, random_state=5)
```

```python
print(X_train.head())
print(Y_train)
```

```python
# 5. Import Decision Tree Classifier and create the object 
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
```

```python
# 6. Train the model by passing the train values
clf.fit(X_train, Y_train)
```

```python
# 7. Predict by passing the test values
y_predict = clf.predict(X_test)
```

```python
# 8. Check teh score
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, y_predict)
```

```python
# To plotting a tree
from sklearn.tree import plot_tree

plot_tree(clf)
```
