```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BostonHousing.csv")
```

```python
df.head()
```

```python
input_features = df.iloc[:,:-1]
output = df.iloc[:,-1:]
```

```python
input_features.head()
```

```python
output.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_features, output, random_state = 2, test_size = 0.2)
```

```python
param_grid = {
    "criterion" : ["absolute_error", "squared_error", "friedman_mse", "poisson"],
    "max_depth" : [3, 4, 5, 6, 7, 8, 9, 10, None],
    "max_features" : [.25, .50, .75, 1]
}
```

```python
from sklearn.tree import DecisionTreeRegressor

regressior = DecisionTreeRegressor()

regressior.fit(X_train, Y_train)
```

```python
from sklearn.model_selection import GridSearchCV

gscv = GridSearchCV(regressior, param_grid = param_grid, cv = 10, n_jobs =1)
```

```python
gscv.fit(X_train, Y_train)
```

```python
gscv.best_estimator_
```

```python
gscv.best_score_
```

```python
gscv.best_params_
```

```python
model = gscv.best_estimator_
```

```python
y_predict = model.predict(X_test)

y_predict
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

MSE = mean_squared_error(Y_test, y_predict)
MAE = mean_absolute_error(Y_test, y_predict)
R2_Score = mean_squared_error(Y_test, y_predict)

print("Mean Squared Error", MSE)
print("Mean Absolute Error", MAE) 
print("R2 Score", R2_Score)
```
