```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
input_features = df.iloc[:,:-1]
labels = df.iloc[:,-1:]
```

```python
input_features.head()
```

```python
labels.head()
```

```python
labels.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_features, labels, random_state = 2, test_size = .2)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth = 3)
```

```python
clf.fit(X_train, Y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
import dtreeviz

viz = dtreeviz(
    clf,
    X_train,
    Y_train,
    target_name="species",
    feature_names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
 
)


viz.view()
```
