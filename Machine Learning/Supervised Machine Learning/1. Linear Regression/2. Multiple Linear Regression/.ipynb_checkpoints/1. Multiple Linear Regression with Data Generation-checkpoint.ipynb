```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
```

```python
input_features, output = make_regression(n_samples = 100, n_features = 2, n_targets=1, n_informative = 2, noise = 50)
```

```python
print(input_features.ndim)
print(output.ndim)
```

```python
print(input_features.shape)
print(output.shape)
```

```python
print(input_features.size)
print(output.size)
```

```python
print(input_features[:5])
print(output[:5])
```

```python
df = pd.DataFrame({"feature1": input_features[:,0],"feature2": input_features[:,1], "target" : output})
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(input_features[:,:-1], input_features[:,-1:], color='blue')

plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_features, output, test_size = 0.2, random_state = 2)
```

```python
print(X_train[:5])
print(Y_train[:5])
print(X_test[:5])
print(Y_test[:5])
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
```

```python
LR.fit(X_train, Y_train)
```

```python
y_predict = LR.predict(X_test)

y_predict
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

R2_Score = r2_score(Y_test, y_predict)
MAE = mean_absolute_error(Y_test, y_predict)
MSE = mean_squared_error(Y_test, y_predict)
RMSE = root_mean_squared_error(Y_test, y_predict)

print("R2 Score: ", R2_Score)
print("Mean Absolute Error: ", MAE)
print("Mean Squared Error: ", MSE)
print("Root Mean Squared Error: ", RMSE)
```
