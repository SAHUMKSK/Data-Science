```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
```

```python
diabetes = load_diabetes()
```

```python
type(diabetes)
```

```python
feature_values = diabetes.data
target_values = diabetes.target
feature_names = diabetes.feature_names
```

```python
print(feature_values)
print(target_values)
print(feature_names)
```

```python
print(feature_values.shape)
print(target_values.shape)
```

```python
df = pd.DataFrame(data = feature_values, columns = feature_names)
```

```python
df["target"] = target_values
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
input_columns = df.iloc[:,:-1]
output = df.iloc[:,-1:]
```

```python
print(input_columns.head())
print(output.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_columns, output, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
```

```python
LR.fit(X_train, Y_train)
```

```python
y_predict = LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

R2_Score = r2_score(Y_test, y_predict)
MAE = mean_absolute_error(Y_test, y_predict)
MSE = mean_squared_error(Y_test, y_predict)
RMSE = root_mean_squared_error(Y_test, y_predict)

print("R2 Score: ", R2_Score)
print("Mean Absolute Error: ", MAE)
print("Mean Squared Error: ", MSE)
print("Root Mean Squared Error: ", RMSE)
```
