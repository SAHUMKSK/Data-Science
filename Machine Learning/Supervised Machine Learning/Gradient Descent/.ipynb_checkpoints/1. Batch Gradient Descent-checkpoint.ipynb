# Batch Gradient Descent

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BostonHousing.csv")

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
input_features = df.iloc[:,:-1]
outpot = df.iloc[:,-1:]

print(input_features[:5])
print(outpot[:5])
```

```python
from sklearn.preprocessing import StandardScaler

Standard = StandardScaler()

scaled_input_features = Standard.fit_transform(input_features)

print(scaled_input_features[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(scaled_input_features, outpot, random_state = 2, test_size = 0.2)

print(X_train[:5])
print(X_test[:5])
print(Y_train[:5])
print(Y_test[:5])
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import SGDRegressor

SGD = SGDRegressor()

SGD.fit(X_train, Y_train)
```

```python
y_predict = SGD.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

print("Mean Squared Error is", mean_squared_error(Y_test, y_predict))
```
