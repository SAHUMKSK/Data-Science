```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
input_features = df.iloc[:,1:-1]
labels = df.iloc[:,-1]
```

```python
print(input_features)
print(labels)
```

```python
# Convert the data into same scale

from sklearn.preprocessing import StandardScaler, MinMaxScaler

standard_scaler = StandardScaler()

scaled_input_data = standard_scaler.fit_transform(input_features)
```

```python
scaled_input_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(scaled_input_data, labels, random_state = 2, test_size = 0.2)
```

```python
print(X_train[:5])
print(X_test[:5])
print(Y_train[:5])
print(Y_test[:5])
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, Y_train)
```

```python
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, y_predict)
```

```python
param_dist = {
    "criterion" : ["entropy", "gini"],
    "max_depth" : [1, 2, 3, 4, 5, 6, 7, None],
}
```

```python
from sklearn.model_selection import GridSearchCV

gscv = GridSearchCV(clf, param_grid = param_dist, cv = 10, n_jobs = -1)
```

```python
gscv.fit(X_train, Y_train)
```

```python
gscv.best_estimator_
```

```python
gscv.best_score_
```
