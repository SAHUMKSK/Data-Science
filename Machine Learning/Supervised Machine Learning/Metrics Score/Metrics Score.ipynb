```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("1. Placement(cgpa and package).csv")

df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (4, 4))
plt.scatter(df[["cgpa"]], df[["package"]])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
input_features = df.iloc[:,:-1]
output = df.iloc[:,-1:]
```

```python
print(input_features[0:5])
print(output[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input_features, output, test_size =0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(df[["cgpa"]], df[["package"]])
plt.plot(X_train, lr.predict(X_train), color = "red")
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

R2_Score = r2_score(Y_test, y_predict) # 1 - (SSR / SSR)
MAE = mean_absolute_error(Y_test, y_predict) # (∑|y - ŷ|)/n
MSE = mean_squared_error(Y_test, y_predict) # ∑(y - ŷ)^2/n
RMSE = root_mean_squared_error(Y_test, y_predict) #  √∑(y - ŷ)^2/n

print("R2 Score is:", R2_Score)
print("Mean Absolute Error is:", MAE)
print("Mean Squared Error is:", MSE)
print("Root Mean Squared Error is:", RMSE)
```
